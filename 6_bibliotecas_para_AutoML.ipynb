- http://epistasislab.github.io/tpot/

```python
%pip install tpot
```

```python
import pandas as pd
import numpy as np

from sklearn.datasets import load_iris
```

```python
iris = load_iris()

print(iris.DESCR)
```

```python
df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)

df_iris['Species'] = iris.target
```

```python
df_iris.head()
```

```python
df_iris.info()
```

```python
print(pipeline_optimizer.score(Xval, yval))
```

#### **4. AutoML with Cleanlab: auto detection of dirty labels**

```python
%pip install -U skorch
```

```python
%pip install cleanlab
```

```python
import torch

from torch import nn

from sklearn.datasets import fetch_openml
from sklearn.model_selection import cross_val_predict as cvp 
from sklearn.metrics import accuracy_score as acc_score 

from skorch import NeuralNetClassifier 
```

```python
mnist = fetch_openml('mnist_784')
```

```python
X = mnist.data.astype('float32').to_numpy()

X.shape
```

```python
X /= 255.0

X = X.reshape(len(X), 1, 28, 28)
```

```python
X.shape
```

```python
labels = mnist.target.astype('int64').to_numpy()
```

```python
class ClassifierModule(nn.Module):
    
    def __init__(self):
        super().__init__()

        self.cnn = nn.Sequential(
            nn.Conv2d(1, 6, 3),
            nn.ReLU(),
            nn.BatchNorm2d(6),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(6, 16, 3),
            nn.ReLU(),
            nn.BatchNorm2d(16),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )

        self.out = nn.Sequential(
            nn.Flatten(),
            nn.LazyLinear(128),
            nn.ReLU(),
            nn.Linear(128, 10),
            nn.Softmax(dim=-1)
        )

    def forward(self, X):
        X = self.cnn(X)
        X = self.out(X)

        return X
```

```python
model_skorch = NeuralNetClassifier(ClassifierModule)

model_skorch
```

```python
num_crossval_folds = 5
```

```python
pred_probs = cvp(
    model_skorch,
    X,
    labels,
    cv=num_crossval_folds,
    method='predict_proba',
    n_jobs=-1
)
```

```python
pred_probs[:5]
```

```python
predicted_labels = pred_probs.argmax(axis=1)
```

```python
predicted_labels[:15]
```

```python
acc = acc_score(labels, predicted_labels)

print(f'Cross validated estimate of acciracy on held-out data: {acc}')
```

```python
from cleanlab.filter import find_label_issues
```

```python
ranked_label_issues = find_label_issues(
    labels,
    pred_probs,
    return_indices_ranked_by='self_confidence'
)
```

```python
len(ranked_label_issues), ranked_label_issues
```

```python
print(f'Cleanlab found {len(ranked_label_issues)} label issues')
print(f'\nTop 15 most likely label errors: \n{ranked_label_issues[:15]}')
```

```python
import matplotlib.pyplot as plt
```

```python
def plot_examples(id_iter, nrows=1, ncols=1):
    plt.figure(figsize=(10, 10))

    for count, id in enumerate(id_iter):
        plt.subplot(nrows, ncols, count + 1)
        plt.imshow(X[id].reshape(28, 28), cmap='gray')
        plt.title(f'id: {id} \n label: {labels[id]}')
        plt.axis('off')

    plt.tight_layout(h_pad=2.0)
```

```python
plot_examples(ranked_label_issues[range(15)], 3, 5)
```

#### **5. AutoML with Imbalanced-Learn: auto balancing of imbalanced data**

```python
from collections import Counter 

from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression 
from sklearn.model_selection import train_test_split as tts
from sklearn.preprocessing import StandardScaler

from imblearn.datasets import make_imbalance
from imblearn.metrics import classification_report_imbalanced
from imblearn.pipeline import make_pipeline
from imblearn.under_sampling import NearMiss

print(__doc__)
```

```python
random_state = 1522
```

```python
iris = load_iris()

X, y = make_imbalance(
    iris.data,
    iris.target,
    sampling_strategy={0: 25, 1: 50, 2: 50},
    random_state=random_state             
)
```

```python
X_train, X_val, y_train, y_val = tts(
    X, y, random_state=random_state, test_size=0.3
)
```

```python
print(f'Training target statistics: {Counter(y_train)}')
print(f'Validation target statistics: {Counter(y_val)}')
```

```python
pipeline = make_pipeline(
    NearMiss(version=2), 
    StandardScaler(),
    LogisticRegression(random_state=random_state, n_jobs=-1)

)

pipeline
```

```python
pipeline.fit(X_train, y_train)
```

```python
print(classification_report_imbalanced(y_val, pipeline.predict(X_val)))
```

```python
import pandas as pd
import numpy as np

from flaml import AutoML

from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split as tts
```

```python
Xtrain, Xval, ytrain, yval = tts(df_wine.values, wine.target,
                                 test_size=0.3, random_state=1522)
```
